```python
import random
options = ['true', 'false']
correct_answer = {
    1:'true',
    2:'true',
    3:'true',
    4:'false',
    5:'true',
    6:'false',
    7:'false',
    8:'false',
    9:'false',
    10:'true'
    }
the_questions = {
    1:'World cup is played every four years.',
    2:'Fuel price in Nigeria is on steady increase.',
    3:'Nigeria flag is green and white.',
    4:'Wunmi is a fat lady.',
    5:'Is python a programming language.',
    6:'Is Gomycode a supermarket?',
    7:'Smokers are liable to live long',
    8:'Is Gomycode in Kano? ',
    9:'Is the color of water red? ',
    10:'Is pluto part of the solar system? '
    }

class Player:
    def __init__(self, name, gender, age, state_of_origin):
        self.name = name
        self.gender = gender
        self.age = age
        self.state_of_origin = state_of_origin
        self.score = 0
        self.attempts = 3
        self.check = set()

    def questions(self):
        while True:
            question_no = random.choice(list(the_questions.keys()))
            if question_no not in self.check:
                self.check.add(question_no)
                print(the_questions[question_no])
                question_ans = correct_answer[question_no]
                answer = input('Enter your answer: ').lower()

                if answer == question_ans:
                    self.score += 1
                    print('Correct!')
                elif answer not in options:
                    self.attempts -= 1
                    print("Invalid input. Please enter 'true' or 'false'.")
                    break

                else:
                    answer != question_ans
                    self.attempts -= 1
                    print('Incorrect answer')
                    break


    def performance(self):
        all_scores = {}
        all_scores[self.name] = self.score
        print(all_scores)




```

```python
player1 = Player('AYO', 'MALE', 21, 'OGUN')

for values in range(10):
    player1.questions()
    if player1.attempts == 0:
      print('Sorry, you have exhausted your number of trials')
      break
```

```python
player1.performance()
player1.score = player1.score
```
