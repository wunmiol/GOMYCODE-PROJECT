```python
#Here we are trying to display our value withouth the index in front,
# so we index out male, this automatically removes the index in front and displays "male"
data1['Gender'].mode()[0]
```

```python
#Having 'Male' as the most occuring gender entry, we assume that the null values should be replaced with 'male'
data1['Gender'].fillna('Male')
```

```python
data1['Gender'].fillna(data1['Gender'].mode()[0])
```

```python
#This is still a method of trying to replace the null values with 'Male', which is the most occuring gender entry
#Note that before this changes reflect in your data set, you have to reassign the entire operation into a variable name indexing the column name
data1['Gender'] = data1['Gender'].fillna(data1['Gender'].mode()[0])
data1['Gender']
```
