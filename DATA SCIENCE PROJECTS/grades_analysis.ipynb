```python
import numpy as np
```

```python
grades = np.array([85, 90, 88, 92, 95, 80, 75, 98, 89, 83])
grades
```

```python
#mean
np.mean(grades)
```

```python
#Median
np.median(grades)
```

```python
#Standard Deviation
np.std(grades)
```

```python
#Maximum Grade
np.max(grades)
```

```python
#Minimum Grade
np.min(grades)
```

```python
#Sorting in Ascending
np.sort(grades)
```

```python
 # index of the highest grade using argmax
np.argmax(grades)
```

```python
#7-Use numpy function to count the number of students who scored above 90.
np.sum(grades > 90)
```

```python
#Use numpy function to calculate the percentage of students who scored above 90.
np.mean(grades > 90) * 100
```

```python
#Use numpy function to calculate the percentage of students who scored below 75.
np.mean(grades < 75) * 100
```

```python
#Use numpy function to extract all the grades above 90 and put them in a new array called "high_performers".
high_performers = grades[grades > 90]
high_performers
```

```python
#Create a new array called "passing_grades" that contains all the grades above 75.
passing_grades = grades[grades > 75]
passing_grades
```

```python
print('Mean:',np.mean(grades),'\n')
print('Median:',np.median(grades),'\n')
print('Standard Deviation:',np.std(grades),'\n')
print('Maximum Grade:',np.max(grades),'\n')
print('Minimum Grade:',np.min(grades),'\n')
print('Ascending:',np.sort(grades),'\n')
print('index of the highest grade:', np.argmax(grades),'\n')
print('Number of students that scored above 90:',np.sum(grades > 90),'\n')
print('Above score 90 percentage:', np.mean(grades > 90) * 100,'%','\n')
print('Below score 75 percentage:', np.mean(grades < 75) * 100,'%','\n')
print('Number of students that scored above 90:',high_performers,'\n')
print('Number of students that scored above 75:',passing_grades,'\n')
```
