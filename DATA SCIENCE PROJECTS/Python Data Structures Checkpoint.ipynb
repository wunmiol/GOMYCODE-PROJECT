```python
shopping_list = []

while True:
    options = input('ENTER AN OPTION:\n~Add, \n~Remove \n~View \n~Exit: ').lower()

    if options == 'add':
        item_to_add = input('What do you want to add? ')
        shopping_list.append(item_to_add)

    elif options == 'remove':
        item_to_remove = input('What do you want to remove? ')
        if item_to_remove in shopping_list:
            shopping_list.remove(item_to_remove)
        else:
            print('Item not found in your cart.')

    elif options == 'view':
        print('Your shopping cart:', shopping_list)

    elif options == 'exit':
        print('Thank you for shopping with us.')
        break

    else:
        print('Please put a valid output.')
```
