1. Hassan has five kinds of dogs which are husky, pitbull, shepherd, bulldog and eskimo.
Create a python program such that anyone can check if Hassan has a particular breed, if the breed is in the list of dogs print 'present' else print 'absent'. Note: use your input and conditional statements.


```python
Hassan_breed = ['husky', 'pitbull', 'shepherd', 'bulldog', 'eskimo']
breed_check = input('Enter breed name: ').lower()

if breed_check in Hassan_breed:
  print('Present')
else:
  print('Absent')

```

3. Create a lambda function that can add two numbers and gives a result.
e.g, addition of 2 and 3 should give 5.
Use a lambda function.



```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))
sum = lambda num1, num2 : num1 + num2
print(f'{num1} + {num2} = {sum(num1, num2)}')
```
