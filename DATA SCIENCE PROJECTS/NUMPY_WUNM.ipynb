Element Wise Properties of Arrays

```python
### Multiplying each element of array_b by 2
array_a = np.array([7,8,9])
array_a

array_b = np.array([[1,2,3],[4,5,6]])
array_b

array_b = array_b * 2
array_b
```

```python
## Since lists don't work elementwise, we're concatenating [2] to list_a.
list_a = [1,2,3]
list_a + [2]
```

```python
## Element Wise addition adds 2 to each element of array_a.
array_a + 2
```

```python
## Element Wise multiplication.
## We multiply each individual element of array_a by its corresponding element in the second row of array_b.
array_a * array_b[1]
```

```python
## The order of the elements matters for elementwise subtraction, division, as well as other operations.
array_b - array_a
```

Arrange function:

In NumPy, the "arange" function is used to create an array with evenly spaced values within a given range. The function is similar to the built-in Python range function, but it returns a NumPy array instead of a list.

```python
list(range(8))
```

```python
## Creates an ndarray with the values in this range.
array_rng = np.arange(30)
array_rng
```

```python
# array_rng = np.arange(stop =  30)
array_rng = np.arange(stop =  30)
array_rng
```

```python
# If we specify only a start, the function assumes this is the "stop" and starts from the origin (0).
array_rng = np.arange(start = 0, stop =  30)
array_rng
```

```python
# "Step" doesn't have to be the same type as the values of the array.
array_rng = np.arange(start = 0, stop =  30, step = 2.5)
array_rng
```

```python
# The casting happens after all the computations.
array_rng = np.arange(start = 0, stop =  30, step = 2.5, dtype = np.float32)
array_rng = np.arange(start = 0, stop =  30, step = 2.5, dtype = np.int32)
array_rng
```

```python
matrix_A = np.array([[1,0,0,3,1],[3,6,6,2,9],[4,5,3,8,0]])
matrix_A
```

```python
np.argmin(matrix_A)
```

```python
#minimum/maximum in the entire array
print("Minimum value using np.min: ",np.min(matrix_A))
print("Maximum value using np.max: ",np.max(matrix_A))

# Find the minimum/maximum value along a specific axis=0 (along columns)
print("Minimum value using np.amin along axis 0: ",np.amin(matrix_A, axis=0))
print("Maximum value using np.amax along axis 0: ",np.amax(matrix_A, axis=0))

# Find the minimum/maximum value along a specific axis=1 (along rows)
print("Minimum value using np.amin along axis 1: ",np.amin(matrix_A, axis=1))
print("Maximum value using np.amax along axis 1: ",np.amax(matrix_A, axis=1))

# Compute the minimum/maximum value along axis 0 (column-wise) using np.minimum.reduce
print("Minimum value using np.minimum.reduce: ",np.minimum.reduce(matrix_A))
print("Maximum value using np.maximum.reduce: ",np.maximum.reduce(matrix_A))
```

Averages and Variances:

```python
matrix_A = np.array([[1,0,0,3,1],[3,6,6,2,9],[4,5,3,8,0]])

# Find the Median value along a specific axis
print("Median value along axis 0: ",np.median(matrix_A, axis=0))
print("Median value along axis 1: ",np.median(matrix_A, axis=1))
print('\n')

# Find the Mean value along a specific axis
print("Mean value along axis 0: ",np.mean(matrix_A, axis=0))
print("Mean value along axis 1: ",np.mean(matrix_A, axis=1))
print('\n')

# Find the Variance value along a specific axis
print("Variance value along axis 0: ",np.var(matrix_A, axis=0))
print("Variance value along axis 1: ",np.var(matrix_A, axis=1))
print('\n')

# Find the Standard Deviation value along a specific axis
print("Standard Deviation value along axis 0: ",np.std(matrix_A, axis=0))
print("Standard Deviation value along axis 1: ",np.std(matrix_A, axis=1))
print('\n')

# Find the Average value along a specific axis
print("Variance value along axis 0: ",np.average(matrix_A, axis=0))
print("Variance value along axis 1: ",np.average(matrix_A, axis=1))
```

Covariance and correlation :

```python
#Covariance
matrix_A = np.array([[1,0,0,3,1],[3,6,6,2,9],[4,5,3,8,0]])
cov_matrix = np.cov(matrix_A)
print("Covariance matrix of matrix_A : \n",cov_matrix)
```

```python
#Correlation
matrix_A = np.array([[1,0,0,3,1],[3,6,6,2,9],[4,5,3,8,0]])
corr_matrix = np.corrcoef(matrix_A)
print("Correlation matrix of matrix_A : \n",corr_matrix)
```
